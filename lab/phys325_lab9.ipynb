# Problem 1: Finding the root of a function & interpolate

In the mean-field theory of ferro-magnetism, the strength of magnetization $m$ is related to temperature (T) via

$$
m = \tanh(C m / T)
$$
We can see that $m=0$ is always a solution, but we are interested in finding $m\neq0$ solutions.

Let's assume C = 1 for simplicity.

(a)  For $T=0.5$, plot $f(m)= m - \tanh(m/T)$, i.e., f as a function of m, for m$\in$[0,1].  Add a horizontal line $f=0$. Where does the curve cross 0?

(b) Use `scipy.optimize.root_scalar` to find the non-zero root $m$ at $T = 0.5$.

(c) Repeat Part (a), for  T = 0.2, 0.4, ..., 1.2, plot $f(m)$. Show all curves together and add the legend.  Above which temperature is there no non-zero solutions (i.e. the material cannot be magnetized)?

(d) Write a function `magnetization(T)` that takes in $T$, find the $m$ that satisfies the equation, and returns this $m$.

(e) Plot $m$ as a function of $T$, for the range that $m\neq0$ solution exists.

(f) Create an interpolation function for m(T), calculate the interpolated value at T = 0.25

# Problem 2: ODEs with two variables: x(t) and y(t)

The Lotka-Volterra equations describe the model of predator-prey interactions. For example, let $x$ represent the population size of rabbits and $y$ represent population size of foxes.
The rabbits reproduce at a rate proportional
to their population, but are eaten by the foxes at a rate proportional to
both their own population and the population of foxes:

$$
\frac{dx}{dt} = \alpha x - \beta xy,
$$
where $\alpha$ and $\beta$ are constants.  At the same time the foxes
reproduce at a rate proportional the rate at which they eat
rabbits, because they need food to grow and reproduce, but also die of
old age at a rate proportional to their own population:

$$
\frac{dy}{dt} = \gamma xy - \delta y,
$$
where $\gamma$ and $\delta$ are also constants.
Greek letters:
$\alpha$: alpha,
$\beta$: beta,
$\gamma$: gamma,
$\delta$: delta.

(a)  Explain in words: what are the meanings of the four parameters $\alpha$, $\beta$, $\gamma$, $\delta$?

(b) Let's set $\alpha=1$, $\beta=\gamma=0.5$, $\delta=2$ , and the initial condition $x=y=2$ at $t=0$.  Solve these equations and plot $x(t)$ and $y(t)$ from $t=0$ to $t=20$.

(c) Repeat the same calculation for $\alpha=0.5$, $\beta=\gamma=0.5$.

(d) Imagine you are a rabbit in this system. Describe the two scenarios, part (b) and part (c), to your kin.
